```python
class Emp():
    def __init__(self,emp_name:str,age:int):
        self.emp_name = emp_name
        self.age = age
```

```python
employee = Emp("Rohan",22)
```

```python
employee.emp_name, employee.age
```

```python
from dataclasses import dataclass
```

```python
@dataclass
class Employee():
    emp_name:str
    age:int
```

```python
emp = Employee(emp_name="Rohan",age=22)
```

```python
emp.emp_name, emp.age
```

```python
drop_col_name=[]
for column_name,missing_prec in zip((data.isna().sum()/data.shape[0]).index, (data.isna().sum()/data.shape[0]).values):
    print(column_name,missing_prec*100)
    if missing_prec > 0.3:
        drop_col_name.append(column_name)
```
